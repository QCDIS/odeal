```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt

def extract_model_name(file_path):
    # Extracts the model name from a file path
    file_name = os.path.basename(file_path)
    return file_name.split('_')[0]

def group_files_by_conditions(directory, comparison_pairs, key_variables):
    grouped_files = {}
    for n_initial, k, split_method, model_name, query_strategy in comparison_pairs: 
    # Groups the files based on their n_initial and k values
        files = glob.glob(os.path.join(directory, query_strategy, f'{model_name}_{split_method}_{n_initial}_initial_{k}_k.csv'))
        key = []
        for v in key_variables: 
            key.append(eval(v))
        key = tuple(key)

        if key not in grouped_files:
            grouped_files[key]=files
    
    return grouped_files

def read_metrics(file_paths, metric):
    # Reads the F1 scores from CSV files
    metrics = []
    
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        metrics.append(df[metric].values)  # Assuming F1 Score is in the first column
        # f1_scores.append(df['F1-score'].values)  
    
    return metrics

def read_num_samples(file_path):
    # Reads the number of samples from a CSV file
    df = pd.read_csv(file_path)
    return df['num_samples'].values # Assuming num_samples is in the first column


def find_first_index_with_differences(numbers, diff_threshold, consecutive_count):
    for i in range(len(numbers) - consecutive_count + 1):
        if numbers[i] < 0.7699: 
            continue
        consecutive_small_diff = True
        for j in range(1, consecutive_count):
            if abs(numbers[i + j] - numbers[i + j - 1]) >= diff_threshold:
                consecutive_small_diff = False
                break

        if consecutive_small_diff:
            return i

    return -1  # Return -1 if no such index is found

def plot_metrics(grouped_files, metric):
    # Plots the F1 scores for different models using the number of samples as the x-axis
    plt.figure(figsize=[5,2])
    for key, files in grouped_files.items():
        print(key)
        metrics = read_metrics(files, metric)

        num_samples = read_num_samples(files[0])  # Assuming all files have the same num_samples
        for value in metrics:
            # Find the point where the model becomes stable 
            stable_index = find_first_index_with_differences(value, 0.001, 10)
            print(f'{metric}: {value[-1].round(4)} Stable point: {stable_index+1}; {value[stable_index].round(4)}')
            plt.plot(num_samples, value, label=key, linewidth=1.5, color=COLOR_MAP[key[1]], linestyle = LINE_MAP[key[0]])
            # plt.plot(num_samples, value, label=key, linewidth=1.5, color=COLOR_MAP[key[1]], linestyle = LINE_MAP[key[0]], marker=MARKER_MAP[key[0]], markerfacecolor='none', markersize=5)

    plt.xlabel('# queried samples')
    plt.ylabel(metric)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.4), ncol=2, fancybox=True)
    # plt.legend(loc='best')
    # plt.grid(True)
    # print(f'========== Initial:{n_initial}   K:{k} ==========')
    
    figure_file = os.path.join(RESULT_PATH, f'{metric}_{purpose}.png')
    plt.savefig(figure_file, dpi=300, bbox_inches='tight')
    plt.show()

# Example usage
print(f'------ {float_number} -------')
directory = RESULT_PATH  # Replace with the actual directory path
grouped_files = group_files_by_conditions(directory, comparison_pairs, key_variables)
plot_metrics(grouped_files, 'F1-score')
plot_metrics(grouped_files, 'Precision')
plot_metrics(grouped_files, 'Recall')
```
